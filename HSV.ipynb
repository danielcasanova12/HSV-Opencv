```python
import cv2
import numpy as np

# Inicializa a captura de vídeo
cap = cv2.VideoCapture(0)

while True:
    # Captura o frame atual da câmera
    _, frame = cap.read()

    # Converte o frame para o espaço de cor HSV
    hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Define o intervalo para a cor vermelha e cria a máscara
    low_red = np.array([161, 155, 84])
    high_red = np.array([179, 255, 255])
    red_mask = cv2.inRange(hsv_frame, low_red, high_red)
    red = cv2.bitwise_and(frame, frame, mask=red_mask)

    # Define o intervalo para a cor azul e cria a máscara
    low_blue = np.array([94, 80, 2])
    high_blue = np.array([126, 255, 255])
    blue_mask = cv2.inRange(hsv_frame, low_blue, high_blue)
    blue = cv2.bitwise_and(frame, frame, mask=blue_mask)

    # Define o intervalo para a cor verde e cria a máscara
    low_green = np.array([25, 52, 72])
    high_green = np.array([102, 255, 255])
    green_mask = cv2.inRange(hsv_frame, low_green, high_green)
    green = cv2.bitwise_and(frame, frame, mask=green_mask)

    # Define o intervalo para todas as cores, exceto branco
    low = np.array([0, 42, 0])
    high = np.array([179, 255, 255])
    mask = cv2.inRange(hsv_frame, low, high)
    result = cv2.bitwise_and(frame, frame, mask=mask)

    # Exibe os frames com as cores detectadas
    cv2.imshow("Frame", frame)
    cv2.imshow("Red", red)
    cv2.imshow("Blue", blue)
    cv2.imshow("Green", green)
    cv2.imshow("Result", result)

    # Sai do loop se a tecla 'ESC' for pressionada
    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    low_red = np.array([161, 155, 84])
    high_red = np.array([179, 255, 255])
    red_mask = cv2.inRange(hsv_frame, low_red, high_red)
    red = cv2.bitwise_and(frame, frame, mask=red_mask)

    # Encontra contornos na máscara vermelha
    contours, _ = cv2.findContours(red_mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    red_count = len(contours)

    # Desenha os contornos no frame original
    cv2.drawContours(frame, contours, -1, (0, 0, 255), 2)
    
    # Exibe a contagem de objetos na janela
    cv2.putText(frame, f'Red Objects: {red_count}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    # Exibe os frames com a cor detectada e contornos
    cv2.imshow("Frame", frame)
    cv2.imshow("Red", red)

    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
